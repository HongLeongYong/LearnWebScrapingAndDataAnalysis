```python
from bs4 import BeautifulSoup
import requests
```

```python
print('蘋果今日焦點')
dom = requests.get('https://tw.appledaily.com/realtime/recommend/').text
soup = BeautifulSoup(dom, 'html5lib')
outputs = soup.find_all('div','text-container box--margin-left-md box--margin-right-md')
count = 0 
for output in outputs:
    count+=1
    print('第'+ str(count) + ':  ' + output.span.text.strip())
    print('      ' + output.find('div','timestamp').text.strip()+'\n')
```

```python
print('自由即時日門新聞')
dom = requests.get('https://news.ltn.com.tw/list/breakingnews/popular').text
soup = BeautifulSoup(dom,'html5lib')
for els in soup.find('ul','list').find_all('li'):
    print(els.find('a','tit').text.strip())
```
